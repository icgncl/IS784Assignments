In order to load last saved models, I have used main directory of output folder

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
 
tokenizer = GPT2Tokenizer.from_pretrained('/content/output_of_model')
model = GPT2LMHeadModel.from_pretrained('/content/output_of_model')
```

In order to greate new text, firstly input data is encoded, then model is used and finally output of model is decoded.

```python
input_ids = tokenizer.encode('Sir, this is certain.',return_tensors='pt')

output_of_model = (model.generate(input_ids, do_sample=True, max_length=200))[0]
shakespare_text = tokenizer.decode(output_of_model, skip_special_tokens=True)

print(shakespare_text)
```
